```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import matplotlib.animation as animation

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
combo_df = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')

# Display the data table for preview
combo_df.head()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
summary_regimen = combo_df.groupby("Drug Regimen")

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary dataframe.

#Might have to use these outputs later
tmean = summary_regimen["Tumor Volume (mm3)"].mean()
tmedian = summary_regimen["Tumor Volume (mm3)"].median()
tvar = summary_regimen["Tumor Volume (mm3)"].var()
tstd = summary_regimen["Tumor Volume (mm3)"].std()
tsem = summary_regimen["Tumor Volume (mm3)"].sem()

#Take the above outputs and create a new df
summary_df = pd.DataFrame([tmean , tmedian , tvar , tstd , tsem])
summary_tumor_table = summary_df.T
summary_tumor_table.columns = ["Mean", "Median", "Variance", "Standard Dev." , "SEM"]

summary_tumor_table
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# Using the aggregation method, produce the same summary statistics in a single line
#Simple way to create a new statistical table with '.agg' by getting aggregations by that column
agg_summary_table = combo_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean", "median" , "var" , "std" , "sem"]})
agg_summary_table
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin
treatment = ['Capomulin' , 'Ramicane' , 'Infubinol' , 'Ceftamin']

# Start by getting the last (greatest) timepoint for each mouse
last_grt_tp = combo_df.groupby("Mouse ID")["Timepoint"].max()
last_df = pd.DataFrame(last_grt_tp)
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
new_combo_df = pd.merge(combo_df , last_df , on =("Mouse ID" , "Timepoint"))

#Used 'on=' both mouse id and timepoint to be able to get the tumor volume for the LAST timepoint
#and the last timepoint alone. No 'how' is needed since inner is default and left key order is fine.
new_combo_df
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens
final_tumor_vol = new_combo_df[new_combo_df['Drug Regimen'].isin(treatment)]
final_tumor_vol.sort_values("Drug Regimen")

#Further filtered df showing only mouse id, drug regimen and tumor volume
clean_final_tumor = final_tumor_vol.drop(columns=['Sex', 'Age_months', 'Weight (g)', 'Timepoint', "Metastatic Sites"])
clean_final_tumor.sort_values('Drug Regimen')
```

```python
#Reformatting data to be able to loop through easier
# Locate the rows which contain mice on each drug and get the tumor volumes
regimen_format = final_tumor_vol.groupby(['Drug Regimen','Mouse ID']).agg(group_tumor_regimen = ('Tumor Volume (mm3)',lambda x: x.iloc[-1]))
    
#Making the columns the drug regimens using stacks
regimen_columns = regimen_format.stack(level = 0).unstack(level = 0)

#now I can make the for loop required below work the way I want it to
regimen_columns
```

```python
#used this variable to reset for each drug the for loop iterated through
i = 0
# Calculate the IQR and quantitatively determine if there are any potential outliers. 
# Determine outliers using upper and lower bounds

#We can see that there was only 1 outlier and it was in the Infubinol regimen
Infubinol = clean_final_tumor.set_index("Drug Regimen")
Infubinol.loc["Infubinol"]


for drug in treatment:
    quartiles = regimen_columns[drug].quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    upperbound = upperq + (1.5*iqr)
    lowerbound = lowerq - (1.5*iqr) 
    if i ==0:
        print(f'- - - - - - - - - - - - - -')
    print(f"IQR for {drug}")
    print(f"The lower quartile of {drug} is: {lowerq}")
    print(f"The upper quartile of {drug} is: {upperq}")
    print(f"The interquartile range of {drug} is: {iqr}")
    print(f"The the median of {drug} is: {quartiles[0.5]} ")
    print(f"Values below {lowerbound} could be outliers.")
    print(f"Values above {upperbound} could be outliers.")
    print(f'- - - - - - - - - - - - - -')
    i +=1
print(f'The only potential outlier was mouse c326 in the Infubinol regimen, with a tumor volume of 36.321346')
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest
# Create empty list to fill with tumor vol data (for plotting)
tumor = []

#Go through the for loop using the treatment list created above
#to append to the container above for plotting
for drug in treatment:
    tumor.append(list(regimen_columns[drug].dropna()))

fig1, vdr = plt.subplots()
vdr.set_title('Final Tumor Volume Across 4 Regimens')
vdr.set_ylabel('Tumor Volume (mm3)')
vdr.boxplot(tumor, "Symbol",flierprops={'marker': 'o', 'markersize': 7, 'markerfacecolor': 'fuchsia'})
plt.xticks([1,2,3,4] , treatment)
plt.savefig("Images/boxplot.PNG")
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
cap_weight = combo_df.loc[combo_df["Drug Regimen"] == "Capomulin",:]

cap_df = cap_weight.groupby('Mouse ID').mean()

#Made this to use down below
x = cap_df["Weight (g)"]
y = cap_df["Tumor Volume (mm3)"]

#Plot the scatter plot
plt.scatter(cap_df["Weight (g)"], cap_df["Tumor Volume (mm3)"] , color = "green" , edgecolors = "blue")

#Adding title,xlabel and ylabel to the scatter plot
plt.title("Mouse weight vs Avg Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")
plt.savefig("Images/scatter.PNG")
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
corr_coef = st.pearsonr(x,y)[0]
print(f' The correlation coefficient of mouse weight and volume is: {corr_coef}')
print("This is interpreted as a high positive correlation between the 2 variables.")
```

```python
#This will give me the variables for my slope and intercept
inter_result = st.linregress(x,y)

#Now I can use these vaiables to create the regression line
intercept = 21.552160532685008
slope = 0.9544396890241049
reg_line = slope*(x)+intercept

#Use above scatter to create the scatter plot
plt.scatter(x,y)
plt.title("Correlation between Weight and Tumor Volume")
plt.xlabel("Weight (g)")
plt.ylabel("Tumor Volume (mm3)")

#Now we must get the actual linear regression line and give it a legend and place it in a good location
plt.plot(x, reg_line, color = "r", label = "y={:.2f}x+{:.2f}".format(slope,intercept))
plt.legend(loc = "lower right")

plt.savefig("Images/reg_corr.PNG")
plt.show()
```
